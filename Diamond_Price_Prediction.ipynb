## Model Training

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import ColumnTransformer

num_pipeline=Pipeline(
    steps=[
        ('scalar',StandardScaler())
    ]
)

cat_pipeline=Pipeline(
    steps=[
        ('ordinalencoder',OrdinalEncoder())
    ]
)

categorical_cols=x.select_dtypes(include='object').columns
numerical_cols=x.select_dtypes(exclude='object').columns

preprocessor=ColumnTransformer(
    [
        
        ('num_pipeline',num_pipeline,numerical_cols),
        ('cat_pipeline',cat_pipeline,categorical_cols)
    ]
)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
# Check your column names match
print("Data columns:", x_train.columns)
print("Numerical columns:", numerical_cols)
print("Categorical columns:", categorical_cols)
```

```python
preprocessor.fit_transform(x_train)
```

```python
preprocessor.transform(x_test)
```

```python
x_train=pd.DataFrame(preprocessor.fit_transform(x_train),columns=preprocessor.get_feature_names_out())
x_test=pd.DataFrame(preprocessor.transform(x_test),columns=preprocessor.get_feature_names_out())
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error

def evaluate_model(model_name,model):
    results={}
    results['model name']=model_name

    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    true,predicted=y_test,y_pred
    results['mae'] = mean_absolute_error(true, predicted)
    results['mse'] = mean_squared_error(true, predicted)
    results['rmse ']= np.sqrt(mean_squared_error(true, predicted))
    results['r2_square'] = r2_score(true, predicted)
    return pd.DataFrame([results])
```

```python
models={
    'LinearRegression':LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'Elasticnet':ElasticNet()
}

evaluation=pd.DataFrame()

for model_name,model in models.items():
    res=evaluate_model(model_name,model)
    evaluation=pd.concat([evaluation,res])

evaluation
```
